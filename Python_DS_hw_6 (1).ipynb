***Был использован датасет california_housing***

```python
# Импорт библиотек
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Загрузка датасета

california_housing = fetch_california_housing()
X = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
y = pd.Series(california_housing.target)
```

```python
# Разбивка данных на тренировочные и тестовые наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Создание и обучение модели линейной регрессии
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
# Вывод пяти первых предсказаний
print("Первые пять предсказаний:")
print(predictions[:5])

# Вычисление коэффициента детерминации R2
r2 = r2_score(y_test, predictions)
print("\nКоэффициент детерминации R2:", r2)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_auc_score
```

```python
# Загружаем датасет
df = pd.read_csv('/content/sample_data/creditcard.zip', compression='zip')
```

```python
# Проверяем баланс классов
print(df['Class'].value_counts(normalize=True))
```

```python
# Проверяем наличие числовых данных и пропусков
df.info()
```

```python
# Удаляем строки с пропущенными значениями, если они есть
df.dropna(inplace=True)
```

```python
# Проверяем, все ли пропуски были удалены
print(df.isnull().sum().max())  # должно быть 0, если нет пропусков
```

```python
# Настраиваем отображение всех столбцов
pd.options.display.max_columns = 100
```

```python
#  Просматриваем первые 10 строк датафрейма
df.head(10)
```

```python
# Создаем датафрейм X, исключая столбец 'Class'
X = df.drop('Class', axis=1)
```

```python
# Создаем Series y из столбца 'Class'
y = df['Class']
```

```python
# Разбиваем данные на тренировочный и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# Параметры поиска
parameters = [{
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}]
```

```python
# Создаем модель GridSearchCV
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                           param_grid=parameters,
                           scoring='roc_auc',
                           cv=3)
```

```python
# Обучаем модель на тренировочном наборе данных
grid_search.fit(X_train, y_train)
```

```python
# Просматриваем параметры лучшей модели
print(f'Лучшие параметры: {grid_search.best_params_}')
```

```python
# Предсказываем вероятности классов
y_pred_proba = grid_search.predict_proba(X_test)[:, 1]
```

```python
# Вычисляем AUC на тестовых данных
auc_test = roc_auc_score(y_test, y_pred_proba)
print(f'AUC на тестовых данных: {auc_test}')
```

```python
# Можно также вычислить AUC на тренировочных данных для сравнения
y_pred_proba_train = grid_search.predict_proba(X_train)[:, 1]
auc_train = roc_auc_score(y_train, y_pred_proba_train)
print(f'AUC на тренировочных данных: {auc_train}')
```

Значение AUC (Area Under the Curve) является метрикой качества модели машинного обучения, которая показывает, насколько хорошо модель способна различать классы. Чем ближе значение AUC к 1, тем лучше модель различает классы, а значение AUC равное 0.5 говорит о случайном классификаторе.

В данном случае:
- AUC на тестовых данных: 0.9462664156037156
- AUC на тренировочных данных: 0.9703527882554751

Эти значения говорят о том, что модель хорошо различает классы как на тренировочных, так и на тестовых данных. Значения AUC близки к 1, что свидетельствует о хорошем качестве модели. Однако, разница между AUC на тренировочных и тестовых данных может указывать на переобучение модели.
